```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.callbacks import StreamingStdOutCallbackHandler

model = ChatOpenAI(
    temperature=0.8,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()],
    model="gpt-3.5-turbo",
)
```

```python
poem_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a master of poem. You create a masterpiece of poem with any given keyword. This time, the keyword given will be one of the programming languages. If the keyword is not a programming language, them do not create poem.",
        ),
        ("human", "I want a poem about {programming_language}."),
    ]
)

poem_chain = poem_prompt | model
```

```python
explain_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are an expert specialized on explaining poem. You find meanings or intentions that the poet embedded in the poem. But if you don't get a proper poem, then notify that you didn't receive any poems.",
        ),
        ("human", "{poem}"),
    ]
)

explain_chain = explain_prompt | model
```

```python
final_chain = {"poem": poem_chain} | explain_chain

final_chain.invoke({"programming_language": "python"})
```
